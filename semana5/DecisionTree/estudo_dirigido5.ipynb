```python
#importando as bibliotecas
import pandas as pd #biblioteca utilizada para o tratamento de dados via dataframes 
import numpy as np #biblioteca utilizada para o tratamento de valores numéricos (vetores e matrizes)
import matplotlib.pyplot as plt #biblioteca utilizada para construir os gráficos
from sklearn import linear_model
from word2number import w2n 
```

# Regressão linear

```python
df = pd.read_csv('/content/candidatos.csv')
df
```

Retirando as linhas nulas do dataset

```python
df['test_score'].fillna(0, inplace = True)
df
```

```python
df.dropna(subset = ['experience'], inplace=True)
df
```

```python
result = []
for i in df['experience']:
  value = w2n.word_to_num(i)
  result.append(value)

result
```

```python
df['experience'] = result
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['test_score', 'interview_score', 'experience']], df['salary'])
```

```python
reg.coef_
```

```python
experience = [[8, 9.0, 7]]
```

```python
reg.predict(experience)
```

# Florestas Aleatórias

```python
from sklearn.datasets import load_digits 
digits = load_digits()
```

```python
dir(digits)
```

```python
import plotly.express as px
for i in range(4):
  fig = px.imshow(digits.images[i])
  fig.show()
```

```python
df_rd = pd.DataFrame(digits.data)
df_rd
```

```python
df_rd['target'] = digits.target
df_rd
```

```python
len(digits.data)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_rd.drop(['target'], axis='columns'), digits.target, test_size=0.3)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

model.verbose = 1

model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, y_predicted)
```

```python
fig = px.imshow(confusion, text_auto=True)
fig.show()
```

# Árvore de Decisão

```python
df_ad = pd.read_csv('/content/decision_tree.csv')
df_ad.head()
```

```python
input = df_ad.drop('salary', axis = 'columns')
target = df_ad['salary']
input
```

```python
from sklearn.preprocessing import LabelEncoder
l_company = LabelEncoder()
l_job = LabelEncoder()
l_degree = LabelEncoder()
```

```python
input['l_company'] = l_company.fit_transform(input['company'])
input['l_job'] = l_job.fit_transform(input['job'])
input['l_degree'] = l_degree.fit_transform(input['degree'])
```

```python
input
```

```python
new_input = input.drop(['company', 'job', 'degree'], axis = 'columns')
new_input
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(new_input, target)
```

```python
model.predict([[2, 2, 1]])
```

```python
model.predict([[1, 1, 1]])
```

```python
model.predict([[0, 1, 0]])
```
