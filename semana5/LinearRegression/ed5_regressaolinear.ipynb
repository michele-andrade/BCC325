```python
#importando as bibliotecas
import pandas as pd #biblioteca utilizada para o tratamento de dados via dataframes 
import numpy as np #biblioteca utilizada para o tratamento de valores numéricos (vetores e matrizes)
import matplotlib.pyplot as plt #biblioteca utilizada para construir os gráficos
from sklearn import linear_model
from word2number import w2n 
```

# Regressão linear

```python
df = pd.read_csv('/content/candidatos.csv')
df
```

Retirando as linhas nulas do dataset

```python
df['test_score'].fillna(0, inplace = True)
df
```

```python
df.dropna(subset = ['experience'], inplace=True)
df
```

```python
result = []
for i in df['experience']:
  value = w2n.word_to_num(i)
  result.append(value)

result
```

```python
df['experience'] = result
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['test_score', 'interview_score', 'experience']], df['salary'])
```

```python
reg.coef_
```

```python
experience = [[8, 9.0, 7]]
```

```python
reg.predict(experience)
```
